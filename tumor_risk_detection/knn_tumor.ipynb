```python
#Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Data Import
data = pd.read_csv("data.csv")
data
```

```python
#Data  Manipulation
data.drop(["id","Unnamed: 32"],axis = 1, inplace = True)
data
m = data[data.diagnosis == "M"]
b = data[data.diagnosis == "B"]

```

```python
#Visualization
plt.scatter(m.radius_mean,m.texture_mean,color="red",label = "Kötü")
plt.scatter(b.radius_mean,b.texture_mean,color="green",label = "İyi")
plt.xlabel("radius_mean")
plt.ylabel("texture_mean")
plt.legend()
plt.show()
```

```python
data.diagnosis = [1 if each  == "M" else 0 for each in data.diagnosis]
y = data.diagnosis.values
x_data = data.drop(["diagnosis"],axis = 1)
x_data
```

```python
#Normalization
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data))
x
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 42)
```

```python
#KNN model
from sklearn.neighbors import  KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 11) # K
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
```

```python
print(" {} nn score: {} ".format(11,knn.score(x_test,y_test)))
```

```python
# find k value
score_list = []
score = []
for each in range(1,100):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))
plt.plot(range(1,100),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
print(max(score_list))
print(score_list.index(max(score_list)) + 1)
```
